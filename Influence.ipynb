```python
import numpy as np
import statsmodels.api as sm
import matplotlib.pyplot as plt
```

```python
x = np.array([10, 15, 20, 25, 50])
y = np.array([10, 15, 20, 25, 30])
```

```python
# Plotting
plt.scatter(x, y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Scatter Plot')
plt.show()
```

```python
# Linear regression model
X = sm.add_constant(x)  # add a constant term to the predictor
model = sm.OLS(y, X).fit()
```

```python
# Plotting the regression line
plt.scatter(x, y)
plt.plot(x, model.predict(X), color='red')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Linear Regression Plot')
plt.show()
```

```python
# Leverage
leverage = model.get_influence().hat_matrix_diag
print(f"Leverage: {leverage}")
```

```python
# Standardized residuals
std_residuals = model.get_influence().resid_studentized_internal
print(f"Standardized Residuals: {std_residuals}")
```

```python
# Cook's distance
cooks_distance = model.get_influence().cooks_distance[0]
print(f"Cook's Distance: {cooks_distance}")
```
